```python
# Columns having Null values

# Adult_Mortality
# Hepatitis B
# BMI
# GDP
# Population      
# Malnourished10_19  
# Malnourished5_9    
# Income_Index       
# Schooling        
# Life_Expectancy

# when a column has skewness between -0.5 to +0.5 we do mean imputation
```

```python
# df.Adult_Mortality.skew() = 0.8401035036015758
# Going to treat null with median
df.Adult_Mortality.fillna(df.Adult_Mortality.median(), inplace=True)
```

```python
# df['Hepatitis B'].skew() = -1.9590611205008508
# Going to treat null with median
df['Hepatitis B'].fillna(df['Hepatitis B'].median(), inplace=True)
```

```python
# df.BMI.skew() = -0.32974872476008965
# Going to treat null with mean
df.BMI.fillna(df.BMI.mean(), inplace=True)
```

```python
# df.GDP.skew() = 2.9015219149175016
# Going to treat null with median
df.GDP.fillna(df.GDP.median(), inplace=True)
```

```python
# df.Population.skew() = 6.091816456377106
# Going to treat null with median
df.Population.fillna(df.Population.median(), inplace=True)
```

```python
# df.Malnourished10_19.skew() = 2.0131578094291016
# Going to treat null with median
df.Malnourished10_19.fillna(df.Malnourished10_19.median(), inplace=True)
```

```python
# df.Malnourished5_9.skew() = 2.019601543358905
# Going to treat null with median
df.Malnourished5_9.fillna(df.Malnourished5_9.median(), inplace=True)
```

```python
# df.Income_Index.skew() = -0.4003139130065144
# Going to treat null with mean
df.Income_Index.fillna(df.Income_Index.mean(), inplace=True)
```

```python
# df.Schooling.skew() = -0.8762936121951204
# Going to treat null with median
df.Schooling.fillna(df.Schooling.median(), inplace=True)
```

```python
# df.Life_Expectancy.skew() =-0.5570055450291043
# Going to treat null with median
df.Life_Expectancy.fillna(df.Life_Expectancy.median(), inplace=True)
```

```python
num_df1 = df1.select_dtypes(include=['int64', 'float64'])
num_df1.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df1_num_sc = pd.DataFrame(sc.fit_transform(num_df1), columns=num_df1.columns) 
df1_num_sc # Scaled Dataframe
```

```python
str_df1 = df1.select_dtypes(include=['object'])
str_df1
```

```python
dummy = pd.get_dummies(str_df1, drop_first=False, dtype=int)
```

```python
df1_pre = pd.concat([dummy,df1_num_sc], axis=1) # Axis 1 represents column wise 
df1_pre.head(5)
```

```python
x = df1_pre.iloc[:,0:-1]
x
```

```python
y=df1_pre.iloc[:,[-1]]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
import statsmodels.api as sm
MLR_model1 = sm.OLS(y_train, x_train).fit()
MLR_model1.summary()
```

```python
y_test_pred = MLR_model1.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print('MSE',mean_squared_error(y_true=y_test.Life_Expectancy, y_pred=y_test_pred))
print('MAE',mean_absolute_error(y_true=y_test.Life_Expectancy, y_pred=y_test_pred))
```

```python
from sklearn.metrics import r2_score
r2_score(y_true=y_test.Life_Expectancy, y_pred=y_test_pred)
```
