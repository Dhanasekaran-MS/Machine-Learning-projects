```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

### Data Collection and Processing

```python
# WE Dont Have Header File in Dataset (Column name)
```

```python
sonar = pd.read_csv(r"sonar data.csv", header=None)
```

```python
sonar.head(2)
```

```python
sonar.shape
```

```python
sonar.describe().T
```

```python
sonar[60].value_counts()
```

The Distribution is Almost Normal
 
 M -> Mine
 R -> Rock

```python
sns.heatmap(sonar.corr(numeric_only=True))
```

```python
# Separating Features and Label

X= sonar.drop(columns=60, axis=1)
Y= sonar[60]
```

```python
X
```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.15, random_state=42, stratify=Y)
# Stratify splits the labels equally in train data and test data
```

```python
X.shape, x_train.shape, x_test.shape
```

### Model LR

```python
model= LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
```

```python
training = accuracy_score(y_train_pred, y_train)
print("Accuracy on Training Data: ",training)
```

```python
y_pred = model.predict(x_test)
```

```python
acc= accuracy_score(y_pred, y_test)
```

```python
print("Accuracy on Training Data: ",acc)
```

```python
# The Training Accuracy and Testing accuracy is same. The Model is learned well (Fitted). No Underfit or Overfit
```

```python
confusion_matrix(y_test, y_pred)
```
