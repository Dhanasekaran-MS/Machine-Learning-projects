```python
# The skewness for a normal distribution is zero, and any symmetric data should have a skewness near zero. 
# Negative values for the skewness indicate data that are skewed left and 
# positive values for the skewness indicate data that are skewed right.
df.bore.skew()
```

```python
sns.distplot(df['bore'])
```

```python
df['stroke'].skew()
# replacing with median
```

```python
df['horsepower'].skew()
# replacing with median right skewed
```

```python
df['peak-rpm'].skew()
# mean
```

```python
df['price'].skew()
# right skewed
```

```python
# bore is normally distributes since skewnesss approaches close to zero --> mean replacement
df['bore'] = df.bore.fillna(df.bore.mean())
df['peak-rpm'] = df['peak-rpm'].fillna(df['peak-rpm'].mean())

# replacing with median
df.stroke = df.stroke.fillna(df['stroke'].median())
df.horsepower = df.horsepower.fillna(df.horsepower.median())
df['price'] = df['price'].fillna(df['price'].median())
```

```python
# Select numeric columns
num_df1 = df1.select_dtypes(include=['int64', 'float64'])#int64,float64
num_df1.head()
```

```python
# Standardize numeric features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df1_num_sc = pd.DataFrame(sc.fit_transform(num_df1), columns=num_df1.columns) 
df1_num_sc # Scaled Dataframe
```

```python
# Select categorical columns
str_df1 = df1.select_dtypes(include=['object'])
str_df1
```

```python
str_df1.describe()
```

```python
# le make clm
```

```python
str_df1_sc = pd.get_dummies(str_df1, drop_first=True, dtype=int)
```

```python
str_df1_sc
```

```python
# Combine standardized numeric data and dummy variables
df1_pre = pd.concat([str_df1_sc,df1_num_sc], axis=1)  # Axis 1 represents column wise 
df1_pre.head(5)
```

```python
# Define X (features) and Y (target)
x = df1_pre.iloc[:,0:-1]
x
```

```python
#Y = df1.iloc[:, [-1]]
y=df1_pre.iloc[:,[-1]]
y
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Decision Tree Regressor

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse=mean_squared_error(y_true=y_test.price,y_pred=y_test_pred)
print("MSE : ", mse)
```
